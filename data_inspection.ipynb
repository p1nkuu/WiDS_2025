### Sex - ADHD

```python
fem_ADHD = training_solution[(training_solution['Sex_F']==1) & (training_solution['ADHD_Outcome']==1)]
fem_ADHD = pd.concat([fem_ADHD, categorical_data,quantitative_data] , axis=1, join="inner")
fem_ADHD = fem_ADHD.loc[:,~fem_ADHD.columns.duplicated()]
print(fem_ADHD.shape)
fem_ADHD.head(5)
```

```python
m_ADHD = training_solution[(training_solution['Sex_F']==0) & (training_solution['ADHD_Outcome']==1)]
m_ADHD = pd.concat([m_ADHD, categorical_data,quantitative_data] , axis=1, join="inner")
m_ADHD = m_ADHD.loc[:,~m_ADHD.columns.duplicated()]
print(m_ADHD.shape)
m_ADHD.head(5)
```

```python
fem_non_ADHD = training_solution[(training_solution['Sex_F']==1) & (training_solution['ADHD_Outcome']==0)]
fem_non_ADHD = pd.concat([fem_non_ADHD, categorical_data,quantitative_data] , axis=1, join="inner")
fem_non_ADHD = fem_non_ADHD.loc[:,~fem_non_ADHD.columns.duplicated()]
print(fem_non_ADHD.shape)
fem_non_ADHD.head(5)
```

```python
m_non_ADHD = training_solution[(training_solution['Sex_F']==0) & (training_solution['ADHD_Outcome']==0)]
m_non_ADHD = pd.concat([m_non_ADHD, categorical_data,quantitative_data] , axis=1, join="inner")
m_non_ADHD = m_non_ADHD.loc[:,~m_non_ADHD.columns.duplicated()]
print(m_non_ADHD.shape)
m_non_ADHD.head(5)
```

```python
q_col = ['EHQ_EHQ_Total', 'ColorVision_CV_Score',
       'APQ_P_APQ_P_CP', 'APQ_P_APQ_P_ID', 'APQ_P_APQ_P_INV',
       'APQ_P_APQ_P_OPD', 'APQ_P_APQ_P_PM', 'APQ_P_APQ_P_PP',
       'SDQ_SDQ_Conduct_Problems', 'SDQ_SDQ_Difficulties_Total',
       'SDQ_SDQ_Emotional_Problems', 'SDQ_SDQ_Externalizing',
       'SDQ_SDQ_Generating_Impact', 'SDQ_SDQ_Hyperactivity',
       'SDQ_SDQ_Internalizing', 'SDQ_SDQ_Peer_Problems', 'SDQ_SDQ_Prosocial',
       'MRI_Track_Age_at_Scan']

result = {}
mean_val = fem_ADHD[q_col].mean()
median_val = fem_ADHD[q_col].mean()
result["fem_ADHD"] = pd.DataFrame({'Mean': mean_val, 'Median': median_val})
mean_val = m_ADHD[q_col].mean()
median_val = m_ADHD[q_col].mean()
result["m_ADHD"] = pd.DataFrame({'Mean': mean_val, 'Median': median_val})
mean_val = fem_non_ADHD[q_col].mean()
median_val = fem_non_ADHD[q_col].mean()
result["fem_non_ADHD"] = pd.DataFrame({'Mean': mean_val, 'Median': median_val})
mean_val = m_non_ADHD[q_col].mean()
median_val = m_non_ADHD[q_col].mean()
result["m_non_ADHD"] = pd.DataFrame({'Mean': mean_val, 'Median': median_val})

summary_df = pd.concat(result, axis=1)
summary_df.head(None)
```

```python
fem_ADHD["Group"] = "fem_ADHD"
m_ADHD["Group"] = "m_ADHD"
fem_non_ADHD["Group"] = "fem_non_ADHD"
m_non_ADHD["Group"] = "m_non_ADHD"

df_combined = pd.concat([fem_ADHD, m_ADHD, fem_non_ADHD, m_non_ADHD])

# Loop through each column in q_col and create a box plot
for col in q_col:
    plt.figure(figsize=(8, 6))
    
    # Create box plot
    sns.boxplot(x="Group", y=col, data=df_combined, palette="Set2")
    
    # Formatting
    plt.title(f"Comparison of {col} Across Groups")
    plt.xlabel("Group")
    plt.ylabel(col)
    plt.xticks(rotation=45)  # Rotate labels for readability
    plt.grid(axis='y', linestyle="--", alpha=0.7)  # Add grid for readability
    
    # Show plot
    plt.show()
```
